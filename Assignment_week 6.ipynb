```python
import numpy as np
import pandas as pd
```

```python
population=pd.read_csv('population_wb.csv')
population
```

```python
gdp=pd.read_csv('gdp.csv')
gdp
```

```python
population = population.drop('Unnamed: 0', axis='columns')
population
```

```python
gdp = gdp.drop('Unnamed: 0', axis='columns')
gdp
```

```python
population_tidy = population.melt(
    id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
    var_name='year',      
    value_name='value'
)
```

```python
population_tidy
```

```python
gdp_tidy = gdp.melt(
    id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
    var_name='year',      
    value_name='value'
)
```

```python
gdp_tidy
```

```python
display (population_tidy,gdp_tidy)
```

```python
df= (population_tidy,gdp_tidy)
```

```python
display (df)
```
